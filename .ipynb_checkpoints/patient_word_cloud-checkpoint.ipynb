```python
random.seed(42)
np.random.seed(42)
wc = WordCloud(
        background_color ='white',
        stopwords=stopwords,height=100,width=100,
        contour_color='green',contour_width=10
)

wc.generate(patient_data)
plt.imshow(wc)
```

```python
random.seed(42)
np.random.seed(42)
wc = WordCloud(
        background_color ='white',
        stopwords=stopwords,height=250,width=250,
        contour_color='green',contour_width=10
)

wc.generate(patient_1_data)
plt.imshow(wc)
```
