## Establish Some Sample Text to make word clouds 

```python
def build_word_cloud(image_file, text):
    
    if image_file == None:
        image_file = 'brain.jpg'
        
    if text == None:
        text = " word cloud is a visual representation of text data where the size of each word corresponds to "
        text = text + "frequency or importance within a given text corpus."
    
    shape = np.array(PIL.Image.open(image_file))

    wc = WordCloud(
            background_color ='white',
            stopwords=stopwords,
            mask = shape,
            height=600,
            width=400,
            contour_color='blue',
            contour_width=5
    )

    wc.generate(text)
    
    return(wc) 

testing_word_cloud = True
if testing_word_cloud:
    wc = build_word_cloud('cat3.jpg', cat_text)
    plt.imshow(wc)
```

```python
testing_word_cloud = True
if testing_word_cloud:
    wc = build_word_cloud('tree.jpg', tree_text)
    plt.imshow(wc) 
```
